```python
# Read Json File

import re
def filter_chars(text):
    ignored_chars = ["\\", "/",  "_", "|","{", "}","[","]","~",">","<"]
    for i in ignored_chars:
        text = text.replace(i, " ")
    text = re.sub(r'\s+', ' ', text)
    return text
    
    
new_dict = []

import json


for i in range(0,6):
    json_file = f"../images/page_number_{i}/results.json"

    with open(json_file, 'r') as f:
        data = json.load(f)

    for i in data:
    #     print("Before ", i["text"] , "After ", filter_chars(i["text"]))

        i["text"] =filter_chars(i["text"])
        i["text"] = i["text"].strip()

        if i["text"] == " ": continue
        if len(i["text"]) == 0: continue
        new_dict.append(i)

        i["display"] =i["display"].replace("//","/")


    #     i["text"] =i["text"].replace("\\","")
    #     i["text"] =i["text"].replace("\\","")
    #     i["text"] =i["text"].replace("{","")
    #     i["text"] =i["text"].replace("}","")
    #     i["text"] =i["text"].replace("[","")
    #     i["text"] =i["text"].replace("]","")


    import json
    with open(json_file, 'w') as f:
        json.dump(new_dict, f)
```
